```python
# Convert data type -> df.astype(float) (don't work with char: '1.2 g')
# df.replace('100 g', 100) -> depends of the exactly values
# Regex + replace

# remove numbers to see only units
units = nutrition.astype(str).replace('[^a-zA-Z]','',regex=True)
```

```python
units.head()
```

```python
units = units.mode()
```

```python
# rename() -> df.rename(columns={'height':'Height (m)'})
# dropna() -> df.dropna(how='any/all', axis=0/1, subset=['height'])

import numpy as np
# remove columns without unit
units = units.replace('', np.nan).dropna(axis=1)
```

```python
mapper = {k:k + '_' + units[k].at[0] for k in units}
mapper
```

```python
nutrition = nutrition.rename(columns=mapper)
nutrition.head()
```

```python
nutrition.replace('[a-zA-Z]', '', regex=True, inplace=True)
nutrition.head()
```

```python
nutrition.dtypes.value_counts()
```

```python
nutrition = nutrition.astype(float)
nutrition.info(verbose=False)
```

```python
# Exercise 2 - remove all food that contains at least one nan
nutrition_nonan = nutrition.dropna()
nutrition_nonan.shape
```

```python
# total deleted
nutrition.shape[0]-nutrition_nonan.shape[0]
```

```python
# Select foods with vitamin C between 20 and 40 mg. Get the one with the minimum calories.
nutrition_nonan[nutrition_nonan.vitamin_c_mg.between(20,40)]['calories'].nsmallest(1)
```

```python
# How many food items have vitamin C levels between 2 and 3 standard deviations above the mean
def score_z(x, mean, std):
    return (x-mean)/std

vitamin_c = nutrition_nonan.vitamin_c_mg
vitamin_c_score = vitamin_c.apply(score_z, mean=vitamin_c.mean(), std=vitamin_c.std())
vitamin_c[(vitamin_c_score >=2) & (vitamin_c_score<=3)].shape
```
