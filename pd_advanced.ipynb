```python
# Exercise
def popularity(x):
    if x < 220: return 'relative unknown'
    elif x < 600: return 'kind of popular'
    elif x < 2000: return 'popular'
    else: return 'super-popular'

popularity_col = player.page_views.apply(popularity)
```

```python
player['popularity'] = popularity_col
player.head()
```

```python
# how many super-popular
player.popularity.value_counts()
```

```python
player[player.popularity == 'super-popular'].name.size
```

```python
# insert(index_col, name_col, series)
player.insert(0, 'Popularity', popularity_col,)
```

```python
# Concat method
names_A = ['Adrian', 'Adan', 'Beatriz', 'Carlos', 'Richard']
names_B = ['Adrian', 'Adan', 'Robert', 'Cris', 'Richard', 'Simon', 'Paul']
ages_A = [32, 34, 31, 24, 25]
ages_B = [32, 34, 31,23, 27, 25, 28]
score_A = [78, 89, 90, 78, 79]
score_B = [67, 77, 86, 87, 84, 90, 64]

df_A = pd.DataFrame({'name': names_A, 'age': ages_A, 'score':score_A})
df_B = pd.DataFrame({'name': names_B, 'age': ages_B, 'score':score_B})

print(df_A.shape, df_B.shape)
```

```python
new_df = pd.concat([df_A, df_B])
new_df
```

```python
new_df = pd.concat([df_A, df_B], ignore_index=True)
new_df
```

```python
new_df = pd.concat([df_A, df_B], ignore_index=True).drop_duplicates(subset=['name'])
new_df
```

```python
# Multi-index
new_df = pd.concat([df_A, df_B], keys=['Team_A', 'Team_B'])
new_df
```
